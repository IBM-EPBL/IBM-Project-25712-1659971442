```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import matplotlib.pyplot as plt
import os
```

```python
train_data = tf.keras.utils.image_dataset_from_directory(
  "/content/drive/MyDrive/Dataset",
  validation_split=0.25,
  subset="training",
  seed=120,
  image_size=(180, 180),
  batch_size=batch_size)

```

```python
val_data_set = tf.keras.utils.image_dataset_from_directory(
 "/content/drive/MyDrive/Dataset",
  validation_split=0.25,
  subset="validation",
  seed=120,
  image_size=(180, 180),
  batch_size=batch_size)
```

```python
class_names = train_data.class_names
```

```python
plt.figure(figsize=(15, 15))
for images, labels in train_data.take(1):
  for i in range(3):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
```

#  **Import the ImageDataGenerator library**

```python
from keras.preprocessing.image import ImageDataGenerator
```

# **Configure ImageDataGenerator class**

```python
train_datagen = ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

#  **Apply ImageDataGenerator Functionality to Trainset and Testset**

```python
x_train = train_datagen.flow_from_directory(r'/content/drive/MyDrive/Dataset', target_size=(64, 64), batch_size=5, color_mode='grayscale', class_mode='categorical')
x_test = test_datagen.flow_from_directory(r'/content/drive/MyDrive/Dataset', target_size=(64, 64), batch_size=5, color_mode='grayscale', class_mode='categorical')
```
